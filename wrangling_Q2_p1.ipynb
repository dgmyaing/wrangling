```python
import pandas as pd
import numpy as np
import seaborn as sns

url = 'https://raw.githubusercontent.com/dgmyaing/wrangling/main/data/airbnb_hw.csv'
df = pd.read_csv(url)

print( df.shape, '\n')
df.head()

price = df['Price']
price.unique()

price = df['Price']
price = price.str.replace(',','') # Replace commas with nothing
print( price.unique() , '\n')
price = pd.to_numeric(price,errors='coerce') # Typecast price to float/numeric
print( price.unique() , '\n')
print( 'Total missing: ', sum( price.isnull() ) )
```
